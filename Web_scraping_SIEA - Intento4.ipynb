```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options

# Configuración correcta para las versiones recientes
options = Options()
service = Service(ChromeDriverManager().install())

# Inicializar el driver
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://siea.midagri.gob.pe/portal/calendario/#")
```

```python
# Seleccionar Maiz Amarillo Duro
dropdown = WebDriverWait(driver, 20).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "span.select2-selection.select2-selection--single"))
)
dropdown.click()

search_box = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "input.select2-search__field"))
)
search_box.clear()
search_box.send_keys("Maiz Amarillo Duro")
time.sleep(1)
search_box.send_keys(Keys.RETURN)

# Hacer click en el botón "cosecha" usando el ID correcto
cosecha_button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, "btnCosecha"))
)
cosecha_button.click()
time.sleep(2)
```

```python
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Esperar que los puntos del gráfico estén presentes
puntos = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "highcharts-point"))
)

# Iterar sobre los puntos y obtener sus tooltips
datos = []
for punto in puntos:
    try:
        # Mover el mouse sobre el punto
        actions = ActionChains(driver)
        actions.move_to_element(punto)
        actions.perform()
        time.sleep(0.5)  # Espera para que el tooltip se muestre
        
        # Intentar obtener el tooltip de diferentes maneras
        tooltip = WebDriverWait(driver, 3).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".highcharts-tooltip text, .highcharts-tooltip-container"))
        )
        datos.append(tooltip.text)
        print(f"Dato encontrado: {tooltip.text}")
    except:
        print(f"No se pudo obtener tooltip para un punto")

print(datos)
```

```python
from selenium.webdriver.common.action_chains import ActionChains

def posicionar_cursor(driver, x=999, y=715):
    try:
        # Resetear posición del cursor antes de mover
        ActionChains(driver).move_by_offset(0, 0).perform()
        
        # Mover a la posición especificada
        ActionChains(driver).move_by_offset(x, y).perform()
        print(f"Cursor posicionado en X={x}, Y={y}")
        
    except Exception as e:
        print(f"Error al mover el cursor: {e}")
```

```python
from selenium.webdriver.common.action_chains import ActionChains

def click_en_posicion(driver, x=999, y=715):
    try:
        # Maximizar ventana para asegurar visibilidad
        driver.maximize_window()

        # Llevar el mouse a la posición deseada y hacer clic
        actions = ActionChains(driver)
        actions.move_by_offset(x, y).click().perform()
        print(f"Clic realizado en X={x}, Y={y}")

        # Importante: después de hacer clic, resetea la posición del cursor para evitar acumulaciones
        actions.move_by_offset(-x, -y).perform()

    except Exception as e:
        print(f"Error al hacer clic: {e}")
```

```python
from selenium.webdriver.common.action_chains import ActionChains
import time

def obtener_tooltip_highcharts(driver, x_abs=999, y_abs=715):
    try:
        # Mover el mouse a la posición (0,0) para resetear su posición
        actions = ActionChains(driver)
        actions.move_to_element(driver.find_element_by_tag_name('body'))
        actions.move_by_offset(-999, -999)  # Mover a una esquina para resetear
        actions.perform()
        
        # Ahora mover a la posición absoluta deseada
        actions = ActionChains(driver)
        actions.move_by_offset(x_abs, y_abs).perform()
        print(f"Mouse movido a la posición X={x_abs}, Y={y_abs}")
        
        # Esperar a que aparezca el tooltip
        time.sleep(1)
        
        # Intentar capturar el tooltip usando JavaScript
        tooltip_text = driver.execute_script("""
            var tooltip = document.querySelector('.highcharts-tooltip text');
            if (!tooltip) {
                tooltip = document.querySelector('.highcharts-tooltip tspan');
            }
            if (!tooltip) {
                tooltip = document.querySelector('.highcharts-tooltip-container');
            }
            return tooltip ? tooltip.textContent : "No se encontró el tooltip";
        """)
        
        print(f"Contenido del tooltip: {tooltip_text}")
        return tooltip_text
        
    except Exception as e:
        print(f"Error al obtener tooltip: {e}")
        return None
```
